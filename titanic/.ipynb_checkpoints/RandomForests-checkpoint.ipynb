```python
initial_train_df.columns
selected_features_df = initial_train_df[['Pclass', 'Sex', 'Age', 'Parch', 'Fare', 'Survived']]
missing_age_rows = float(len(selected_features_df[pd.isnull(selected_features_df['Age'])]))
len(selected_features_df)
selected_features_df = selected_features_df.dropna(subset=['Age'],how='any',axis = 0)

selected_features_df.columns[pd.isnull(selected_features_df).any()]
#type(selected_features_df['Age'][0])
#selected_features_df = selected_features_df[selected_features_df['Age'] != '?']
```

**One hot encoding**

```python
df_train = pd.get_dummies(selected_features_df, columns = ['Sex'])
df_train
```

```python
predictors = list(df_train.columns)
predictors.remove('Survived')
predictors
x = df_train[predictors].values
y = list(df_train['Survived'].values)
x
```

```python
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier

max_score = 0
max_estimators = 0
for i in range(100,501,50):
    alg = RandomForestClassifier(n_estimators = i, max_features = 'auto', warm_start = True)
    scores = cross_validation.cross_val_score(alg,x,y,cv=10)
    print "The cross validation accuracy is ", scores.mean()
    if scores.mean() > max_score:
        max_score = scores.mean()
        max_estimators = i
max_estimators
```
